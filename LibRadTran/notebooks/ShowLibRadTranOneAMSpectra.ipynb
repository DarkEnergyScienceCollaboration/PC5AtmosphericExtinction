```python
airmassesindexes=np.arange(10,21)
```

```python
plt.figure(figsize=(15,10))
for filename in filelist:
    fullpath=os.path.join(rootpath,filename)
    data = np.loadtxt(fullpath)
    x=data[:,0]
    y=data[:,1]
    plt.plot(x,y)
    
plt.xlim(0,1000)
plt.ylim(0.6,1.)
plt.xlabel('$\lambda$')
    
```
